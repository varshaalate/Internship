Import Libraries & Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


mart = pd.read_excel("D:/MSP/supermarket_sales.xlsx")
mart.head()
```

Create a basic STRIP PLOT on one NUMERIC variable e.g. Total

```python
sns.stripplot(data=mart, y='Total')
```

Expand markers in STRIP PLOT using jitter

```python
sns.stripplot(data=mart, y='Total',jitter=5)
```

Draw line around the points using LINEWIDTH

```python
sns.stripplot(data=mart, y='Total',linewidth=1,jitter=4)
```

Include third categorical variable with HUE

```python
sns.stripplot(data=mart, y='Total', x='Payment',hue='Gender')
```

Separate each level of hue using DODGE

```python
sns.stripplot(data=mart, y='Total',x='Payment',hue='Gender',jitter=0.2,linewidth=0.5,dodge=True)
```

Draw the strips on top of VIOLIN PLOT

```python
sns.stripplot(data=mart, y='Total',x='Payment',jitter=0.1)
sns.violinplot(data=mart, y='Total',x='Payment',color='gray')
```

 Draw the strips on top of BOX PLOT

```python
sns.stripplot(data=mart, y='Total',x='Payment',color="black")
sns.boxplot(data=mart, y='Total',x='Payment',color='gray')
```
